```python
import numpy as np
import graphviz
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
breast_cancer = load_breast_cancer()
```

```python
type(breast_cancer)
```

```python
breast_cancer
```

```python
# Train the classifier
X = breast_cancer.data
y = breast_cancer.target

clf = DecisionTreeClassifier(random_state=42)
clf = clf.fit(X, y)
```

```python
# Visualize the decision tree
data = tree.export_graphviz(clf,out_file=None,
                              feature_names = breast_cancer.feature_names,
                              class_names = breast_cancer.target_names,
                              filled=True, rounded=True,
                              special_characters=True)

graph = graphviz.Source(data)
graph.render("breast_cancer")
```

```python
graph
```

```python
# Make a prediction
new_data = [[5.0,3.0,1.0,2.0,4.0,5.0,3.0,1.0,2.0,4.0,5.0,3.0,1.0,2.0,4.0,5.0,3.0,1.0,2.0,4.0,5.0,3.0,1.0,2.0,4.0,5.0,3.0,1.0,2.0,4.0]]
predicted_class = clf.predict(new_data)
print(f"The predicted breast cancer diagnosis is: {breast_cancer.target_names[predicted_class]}")
```

```python
# Generate random new data
np.random.seed(42) # For reproducibility
new_data_1 = np.random.rand(1, 30)*breast_cancer.data.max(axis=0)
```

```python
predicted_class_1 = clf.predict(new_data_1)
print(f'The Random New Data is : {new_data_1}')
print(f"The predicted breast cancer diagnosis is: {breast_cancer.target_names[predicted_class_1]}")
```
